```python
# Step 1: Import required Python libraries

import random               # to generate fake sensor values
import time                 # to create delay between readings
import pandas as pd         # to store data in a DataFrame
import matplotlib.pyplot as plt   # to make graphs
from datetime import datetime      # to add timestamps
```

```python
# Step 2: Function to simulate one sensor reading

def generate_sensor_data():
    """
    This function simulates readings from 3 sensors:
    - Temperature
    - Humidity
    - Light intensity
    It returns all data in a dictionary.
    """

    data = {
        'timestamp': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        'temperature_c': round(random.uniform(20.0, 35.0), 2),
        'humidity_percent': round(random.uniform(30.0, 90.0), 2),
        'light_lux': round(random.uniform(100, 1000), 2)
    }

    return data

# Test the function
print("Sample sensor reading:")
print(generate_sensor_data())
```

```python
# Step 3: Generate a continuous stream of sensor readings

print("Generating 5 sensor readings (1 per second):")
print("-" * 70)

for i in range(5):
    sensor_data = generate_sensor_data()
    print(f"Reading {i+1}: {sensor_data}")
    time.sleep(1)   # wait 1 second between readings
```

```python
# Step 4: Collect multiple sensor readings and store in a DataFrame

data_stream = []   # empty list to store all readings

print("Collecting 20 sensor readings...")
for i in range(20):
    reading = generate_sensor_data()
    data_stream.append(reading)
    time.sleep(0.2)  # 200ms gap (faster stream)

# Convert to DataFrame
df = pd.DataFrame(data_stream)

print("\nDataFrame created successfully!")
print(f"Shape: {df.shape} (rows, columns)")

print("\nFirst 5 rows:")
print(df.head())

print("\nDescriptive statistics:")
print(df.describe())
```

```python
# Step 5: Visualize the sensor data

plt.figure(figsize=(12, 8))

# Plot 1: Temperature
plt.subplot(3, 1, 1)
plt.plot(df['temperature_c'], marker='o', label='Temperature (°C)')
plt.ylabel("Temperature (°C)")
plt.title("Simulated Sensor Data Stream")
plt.grid(True)
plt.legend()

# Plot 2: Humidity
plt.subplot(3, 1, 2)
plt.plot(df['humidity_percent'], marker='s', label='Humidity (%)')
plt.ylabel("Humidity (%)")
plt.grid(True)
plt.legend()

# Plot 3: Light
plt.subplot(3, 1, 3)
plt.plot(df['light_lux'], marker='^', label='Light (lux)')
plt.xlabel("Reading Number")
plt.ylabel("Light (lux)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Step 6: Save the DataFrame to a CSV file

filename = "simulated_sensor_data.csv"
df.to_csv(filename, index=False)

print(f"✅ Data saved to '{filename}'")
print(f"📌 Total rows saved: {len(df)}")
print(f"📋 Columns: {list(df.columns)}")
```

```python
# Bonus: Basic data analysis of sensor readings

print("📊 Data Analysis")
print("=" * 50)

# Temperature stats
print("\n🌡 Temperature (°C):")
print(f"• Min:  {df['temperature_c'].min()} °C")
print(f"• Max:  {df['temperature_c'].max()} °C")
print(f"• Mean: {df['temperature_c'].mean():.2f} °C")

# Humidity stats
print("\n💧 Humidity (%):")
print(f"• Min:  {df['humidity_percent'].min()} %")
print(f"• Max:  {df['humidity_percent'].max()} %")
print(f"• Mean: {df['humidity_percent'].mean():.2f} %")

# Light stats
print("\n💡 Light (lux):")
print(f"• Min:  {df['light_lux'].min()} lux")
print(f"• Max:  {df['light_lux'].max()} lux")
print(f"• Mean: {df['light_lux'].mean():.2f} lux")
```
